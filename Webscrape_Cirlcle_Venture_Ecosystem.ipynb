```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://www.circle.com/en/ventures-ecosystem'
page = requests.get(URL)

print(page.text)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(id = 'hs_cos_wrapper_widget_1652388055723')
```

```python
venture_title = results.find_all('div', class_='ventures-card-title')
print(venture_title)
```

```python
title_list = []
for title in venture_title:
    print(title.text, end = '\n')
    title_list.append(title.text)
    
print(len(venture_title))
print((title_list))
```

```python
venture_subtitle = results.find_all('div', class_='ventures-card-subtitle')
print(venture_subtitle)
```

```python
subtitle_list = []
for subtitle in venture_subtitle:
    print(subtitle.text, end = '\n'*2)
    subtitle_list.append(subtitle.text)

print(len(venture_subtitle))
print(subtitle_list)
```

```python
links = results.find_all('a')
links
```

```python
links_updated = links[1:]
link_list = []
for link in links_updated:
    link_url = link['href']
    print(link_url, end = '\n')
    link_list.append(link_url)

print(len(links_updated))
print(link_list)
```

```python
df = pd.DataFrame(data = title_list, columns = ['Title'])
```

```python
df['Subtitle'] = subtitle_list
```

```python
df['Links'] = link_list
```

```python
df.to_csv('venture_ecosystem.csv')
```
