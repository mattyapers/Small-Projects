```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = 'https://www.circle.com/en/customer-stories'
page = requests.get(URL)

print(page.text)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(class_ = 'body-container-wrapper customer-stories-index')
results
```

```python
links = results.find_all('a')
links
```

```python
links_updated = links[1:-2]
link_list = []
for link in links_updated:
    link_url = link['href']
    print(link_url, end = '\n')
    link_list.append(link_url)

print(len(links_updated))
print(link_list)
```

```python
blog_title_list = []

blog_title = results.find('h3', class_ = 'featured-blog-title')
new_blog_title = blog_title.text.strip()

blog_title_list.append(new_blog_title)
```

```python
blog_title_1 = results.find_all('span', class_ = 'blog-post-card-name intro')
test = [
    span_element.parent for span_element in blog_title_1
]
# test
```

```python
for blog in test:
    title = blog.find('span', class_ = 'blog-post-card-name intro')
    print(title.text.strip())
    blog_title_list.append(title.text.strip())
```

```python
len(blog_title_list)
```

```python
df = pd.DataFrame(data = link_list, columns = ['Links'])
```

```python
df.insert(0, 'Description', blog_title_list)
```

```python
df.drop(columns = 'Company_Name')
```

```python
df.to_csv('Customer_Stories.csv')
```
